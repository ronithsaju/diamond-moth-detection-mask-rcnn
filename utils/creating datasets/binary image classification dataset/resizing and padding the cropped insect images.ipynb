```python
os.makedirs('Resized with padding binary image classification dataset v2')
os.makedirs('Resized with padding binary image classification dataset v2/DBM')
os.makedirs('Resized with padding binary image classification dataset v2/non-DBM')
```

```python
for filename in os.listdir("binary image classification dataset v2/DBM"):

  DBM_path = "binary image classification dataset v2/DBM/" + str(filename)
  DBM_image = plt.imread(DBM_path)

  #plt.figure()
  #plt.imshow(DBM_image)
  #plt.show()

  image, window, scale, padding, _ = utils.resize_image(DBM_image, 
                                                        min_dim=150, 
                                                        max_dim=150, 
                                                        mode="square")
  
  #plt.figure()
  #plt.imshow(image)
  #plt.show()

  plt.imsave('Resized with padding binary image classification dataset v2/DBM/%s' % str(filename), image)
```

```python
for filename in os.listdir("binary image classification dataset v2/non-DBM"):

  nonDBM_path = "binary image classification dataset v2/non-DBM/" + str(filename)
  nonDBM_image = plt.imread(nonDBM_path)

  #plt.figure()
  #plt.imshow(nonDBM_image) 
  #plt.show()

  image, window, scale, padding, _ = utils.resize_image(nonDBM_image, 
                                                        min_dim=150, 
                                                        max_dim=150, 
                                                        mode="square")
  
  #plt.figure()
  #plt.imshow(image) 
  #plt.show()

  plt.imsave('Resized with padding binary image classification dataset v2/non-DBM/%s' % str(filename), image)
```
